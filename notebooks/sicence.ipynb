```python
def all_l2_norm(constraints):
    states = [x[0] for x in constraints]
    all_dist = []
    for i, x1 in enumerate(states):
        for j, x2  in enumerate(states[i+1:]):
            d=np.linalg.norm(np.subtract(x1,x2))
            if d - 0 < 1e-2:
                print("0 dist at state %s with action %s and %s" %(str(x1), constraints[i][1], constraints[j][1]))
            all_dist.append(d)
    return sorted(all_dist)
```

```python
path = "results/HalfCheetah-v2/05_18_00_29/"
constraints_p = [f for f in os.listdir(path) if ("constraints.p" in f) ]
```

```python
for fname in constraints_p:
    c = pickle.load(open(path+fname,'rb'))
    #everything = c["all"]
    constraints = c["constraints"]
    print([x[-1] for x in constraints])
    print(all_l2_norm(constraints)[:10])
```

## Animating envs

```python
def display_frames_as_gif(frames):
    """
    Displays a list of frames as a gif, with controls
    """
    plt.figure(figsize=(frames[0].shape[1] / 72.0, frames[0].shape[0] / 72.0), dpi = 72)
    patch = plt.imshow(frames[0])
    plt.axis('off')

    def animate(i):
        patch.set_data(frames[i])

    anim = animation.FuncAnimation(plt.gcf(), animate, frames = len(frames), interval=50)
    display(display_animation(anim, default_mode='loop'))
```
