```python
import matplotlib.pyplot as plt
```

```python
def readfile_(fname):
    all_rewards = []
    with open(fname, 'r') as file:
        current_episode = "1"
        rewards = []
        for line in file:
            fields = line.split("\t")
            episode = fields[0].split(" ")[1]
            if episode != current_episode:
                all_rewards.append(rewards)
                current_episode = episode
                rewards = []
            rewards.append(float(fields[-1].split(" ")[-1].strip()))
    return all_rewards
```

```python
from functools import reduce # Python3
from operator import add

def draw(all_rewards):
    x = [[i]*len(all_rewards[i]) for i in range(len(all_rewards))]
    y = reduce(add , all_rewards)
    plt.scatter(x, y, s=10)
    # Set chart title.
    plt.title("Hopper-v2 learning by imitation")
    # Set x, y label text.
    plt.xlabel("Episode")
    plt.ylabel("branch evaluation reward")
    plt.show()
```

```python
def readfile(fname):
    rewards = []
    with open(fname, 'r') as f:
        for line in f:
            fields = line.split("\t")
            print(fields)
            if len(fields)>2:
                rewards.append(float(fields[1].split(" ")[2]))
    return rewards
```

```python
def readfile_new(fname):
    rewards = []
    with open(fname, 'r') as file:
        for line in file:
            fields = line.split("\t")
            for f in fields:
                if ("Eval" in f) and ("Episode" in f):
                    rewards.append(float(f.split(" ")[2][:-7]))
    return rewards
```
