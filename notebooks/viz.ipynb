```python
import matplotlib.pyplot as plt
```

```python
def readfile(fname):
    rewards = []
    with open(fname, 'r') as f:
        for line in f:
            fields = line.split("\t")
            print(fields)
            if len(fields)>2:
                rewards.append(float(fields[1].split(" ")[2]))
    return rewards
```

```python
def readfile_new(fname):
    rewards = []
    with open(fname, 'r') as file:
        for line in file:
            fields = line.split("\t")
            for f in fields:
                if ("Eval" in f) and ("Episode" in f):
                    rewards.append(float(f.split(" ")[2][:-7]))
    return rewards
```
